```python
import pandas as pd
import numpy as np
import pickle
import time
```

# Creating DataSet

### OD Estimacion

```python
odData = pd.read_excel('./data/OD2007Estimacion2014.xlsx')
odData.dropna(inplace=True)
odData.set_index(['Origen', 'Destino'], inplace=True)
odData
```

```python
vehiculo = [x for x in odData.columns if 'Auto' in x or 'Camioneta' in x]
vehiculo
```

```python
odData.insert(5, 'Vehiculo', odData[vehiculo].sum(axis=1))
odData
```

```python
odData.drop(vehiculo, axis = 1, inplace=True)
odData
```

### Generated Data

```python
fullData = pd.read_csv('./data/fullData.csv', index_col='CODIGO_MZ')
fullData
```

```python
indexes = [x for x in fullData.columns if 'ID' in x or 'Shape' in x or 'Zonificacion' in x or 'UbicaciÃ³n' in x]
```

```python
fD = fullData.drop(indexes, axis = 1)
fD
```

```python
fD.sort_index()
```

#### Adding Prediction Data

```python
with open('./model/origen.pkl', 'rb') as f:
    originModel = pickle.load(f)

with open('./model/destino.pkl', 'rb') as f:
    destinationModel = pickle.load(f)
```

```python
toPredictData = pd.read_csv('./data/selectedDataV2.csv', index_col='CODIGO_MZ')
toPredictData.drop([x for x in toPredictData if 'datosAgrupados' in x], axis = 1, inplace=True)
toPredictData
```

```python
toJoin = toPredictData.copy()
toJoin['Viajes Origen'] = originModel.predict(toPredictData)
toJoin['Viajes Destino'] = destinationModel.predict(toPredictData)
toJoin['Viajes Origen'] = np.int64(round(toJoin['Viajes Origen'], 0))
toJoin['Viajes Destino'] = np.int64(round(toJoin['Viajes Destino'], 0))
toJoin
```

```python
toJoin.drop(toPredictData.columns, axis = 1, inplace = True)
toJoin
```

#### Joining Data

```python
joining = fD.copy()
joining = joining.join(toJoin)
joining
```

```python
odcopy = odData.copy()
odcopy = odcopy.join(joining, on='Origen')
odcopy = odcopy.join(joining, on='Destino', rsuffix='__DESTINO', lsuffix='__ORIGEN')
odcopy
```

### Network Data

```python
import routingpy as rp
```

```python
ors = rp.routers.ORS('5b3ce3597851110001cf6248189853a17c424c3183f8cdf2e74c91df')
ors
```

```python
xy = pd.read_csv('./data/Zonificacion/zonificacionXY.csv', index_col='CODIGO_MZ')
xy.drop('OID_', axis = 1, inplace=True)
xy['list'] = xy[['PX', 'PY']].values.tolist()
xy
```

#### Car

```python
durationData = pd.DataFrame(columns=xy.index, index=xy.index)
durationData
```

```python
for i in range(0,68):
    dest = [i for i in range(7*i, 7*(i+1))]
    if i == 40:
        time.sleep(60)
    dist = ors.matrix(xy.list.values.tolist(), profile = 'driving-car', sources=dest)
    durationData.iloc[dest] = np.array(dist.raw['durations'])

# Se debe correr una vez mÃ¡s para cubrir todos
dest = [476, 477]
dist = ors.matrix(xy.list.values.tolist(), profile = 'driving-car', sources=dest)
durationData.iloc[dest] = np.array(dist.raw['durations'])
```

#### Join

```python
dD = durationData.reset_index().melt(id_vars='CODIGO_MZ', var_name='Destino', value_name='travel_time').rename(columns={'CODIGO_MZ': 'Origen'}).set_index(['Origen', 'Destino']).sort_index()
dD2 = durationData2.reset_index().melt(id_vars='CODIGO_MZ', var_name='Destino', value_name='travel_time').rename(columns={'CODIGO_MZ': 'Origen'}).set_index(['Origen', 'Destino']).sort_index()
```

```python
travelData = dD.join(dD2, lsuffix='_Driving', rsuffix='_Walking')
travelData
```

```python
odcopy = odcopy.join(travelData)
odcopy
```

```python
odcopy.fillna(0, inplace=True)
```
